```python
def fitness(population, n):
  fitness = []   # store the fitness score of each of the individual transactions in the population

  for p in population:
    balance = 0
    for i in range(n):
      if p[i] == 1:
        transaction_type = transaction[i].split()[0]
        transaction_amount = int(transaction[i].split()[1])
        if transaction_type == "d":
          balance = balance + transaction_amount
        else:
          balance = balance - transaction_amount

    fitness.append(abs(balance))

  return fitness
```

```python
def select(population, fit):  #fit = fitness values of all transactions

  weight = [i / np.sum(fit) for i in fit]  #weight ~ probability of each individual being selected.
  select1, select2 = np.random.choice(len(population), 2, weight)  #select parent chromosomes

  return population[select1], population[select2]
```

```python
def crossover(x, y):

  crossover_point = np.random.randint(0,n)
  child = np.concatenate((x[:crossover_point], y[crossover_point:]))   #forming child chromosome

  return child
```

```python
def mutate(child):

  gene = np.random.randint(0, len(child)) #gene index
  if child[gene] == 0:
      child[gene] = 1
  else:
      child[gene] = 0

  return child  #mutated child
```

```python
def GA(population, n, mutation_threshold):
  limit = 1000

  for i in range(limit):
    temp_population = []
    fit = fitness(population, n)

    for j in range(len(population)):
      x, y = select(population, fit)
      child = crossover(x, y)

      if (np.random.random()) < mutation_threshold:
        child = mutate(child)

      if all(i == 0 for i in child):
        temp_population.append(child)
        continue

      if fitness([child], n)[0] == 0:
        c = ''
        for i in child:
          c = c + str(i)
        return c

      temp_population.append(child)
    population = temp_population

  return -1
```
