```python
# CSE422 Lab 1

#input_text = open('input.txt'
from google.colab import drive
drive.mount('/content/drive')
input_text = open('/content/drive/MyDrive/CSE422 L1 input.txt')

grp = {}
heu = {}
while True:
  input_line = input_text.readline()
  if input_line == "":
    break
  input_line = input_line.split()
  city = input_line[0]
  heu2 = input_line[1]
  #print(input_line)
  #print(city, heu2)

  for i in range(2,len(input_line),2):
    u = input_line[i]
    v = int(input_line[i+1])
    #print(u,v)
    if city not in grp:
      grp[city] = []
      heu[city] = int(heu2)   #heuristic of each city
    grp[city].append((u,v))  #adjacent cities and distance

print(grp)
print(heu)
```

```python
import math
import queue
qu = queue.PriorityQueue()
dis = {}   #path cost
h_dis = {}  #path cost + heu
p_node = {}

def A_Star_Search(grp,start,goal):
  for v in grp:
    dis[v] = math.inf     #distance travelled from start
    h_dis[v] = math.inf   #estimated distance to finish from curent city
    p_node[v] = None
  dis[start] = 0
  h_dis[start] = heu[start]
  qu.put((h_dis[start],start))
  #print(qu.qsize())

  while qu.qsize() != 0:
    des = qu.get()
    #print("Destination:", des)
    visited = des[1]
    #print("Visited:",visited)
    for i in grp[visited]:          #city to be visited from the current destination and their distance/path cost
      #print(i)
      next_vist = i[0]
      next_vist_dis = i[1]
      #print(visited, dis[visited])                          #visited city , path cost
      #print(next_vist, dis[visited] + next_vist_dis)        #next visited city , path cost
      #print(dis[next_vist])

      if dis[next_vist] > (dis[visited] + next_vist_dis) :
        dis[next_vist] = dis[visited] + next_vist_dis
        h_dis[next_vist] = dis[next_vist] + heu[next_vist]   #f(n) updated
        #print(next_vist, dis[next_vist], h_dis[next_vist])
        p_node[next_vist] = des[1]
        #print(p_node)
        qu.put((h_dis[next_vist], next_vist))

      if des == goal:
        break

  if dis[goal] == math.inf:
     print("No path found")

  else:
      reached = goal
      path = ""
      while reached != start :
          if reached == goal :
            path = reached + path
            #print(path)
          else:
            path = reached + " -> " + path
            #print(path)

          reached = p_node[reached]
          #print(reached)

      path = start + " -> " + path

      if path == "":
          print("No path found")

      else:
         print("Path: "+ path)
         print("Total Distance : " + str(dis[goal])+" km")

A_Star_Search(grp,'Arad','Bucharest')

#A_Star_Search(grp,'Arad','Neamt')
#A_Star_Search(grp,'Arad','Dobreta')
#A_Star_Search(grp,'Arad','Eforie')
```
